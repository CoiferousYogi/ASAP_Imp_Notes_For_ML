### Data Visualization
- **Histogram**: Useful for visualizing the distribution of numerical data.
- **Box Plot**: Great for showing the spread and skewness of data.
- **Scatter Plot**: Ideal for investigating relationships between two continuous variables.
- **Bar Plot**: Effective for comparing quantities across different categories.
- **Pie Chart**: Good for showing proportional data.

### 1. Histogram
A histogram is used to represent the distribution of numerical data.

```python
import matplotlib.pyplot as plt
import numpy as np

# Example data
data = np.random.randn(1000)

# Plotting the histogram
plt.hist(data, bins=30, edgecolor='black')
plt.title('Histogram')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

### 2. Box Plot
A box plot (or box-and-whisker plot) shows the distribution of data based on a five-number summary.

```python
import matplotlib.pyplot as plt
import numpy as np

# Example data
data = [np.random.randn(100) for i in range(5)]

# Plotting the box plot
plt.boxplot(data)
plt.title('Box Plot')
plt.xlabel('Category')
plt.ylabel('Value')
plt.show()
```

### 3. Scatter Plot
A scatter plot is used to display values for typically two variables for a set of data.

```python
import matplotlib.pyplot as plt

# Example data
x = np.random.rand(100)
y = np.random.rand(100)

# Plotting the scatter plot
plt.scatter(x, y, alpha=0.5)
plt.title('Scatter Plot')
plt.xlabel('X Value')
plt.ylabel('Y Value')
plt.show()
```

### 4. Bar Plot
A bar plot is used to represent categorical data with rectangular bars.

```python
import matplotlib.pyplot as plt

# Example data
categories = ['A', 'B', 'C', 'D']
values = [10, 20, 15, 25]

# Plotting the bar plot
plt.bar(categories, values, color='skyblue')
plt.title('Bar Plot')
plt.xlabel('Category')
plt.ylabel('Value')
plt.show()
```

### 5. Pie Chart
A pie chart is used to represent parts of a whole.

```python
import matplotlib.pyplot as plt

# Example data
labels = ['A', 'B', 'C', 'D']
sizes = [15, 30, 45, 10]
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']

# Plotting the pie chart
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True, startangle=140)
plt.title('Pie Chart')
plt.show()
```
